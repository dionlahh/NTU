# Baysian Model
---
### Why Bayesian Model instead of regular Linear Regression?

**Bayesian analysis is more accurate with smaller samples.**

With our limited dataset size due to the limits of our API keys, we figured that using Bayesian Model would be more suited to us.

Other reasons include:
- Bayesian models more ï¬‚exible, handles more complex models.
- Bayesian hierarchical models easier to extend to many levels.
- Philosophical differences (compared to frequentist analysis).

# Defining Functions

We define 2 functions to help us to obtain the models for different predictors and response variables.

1 for a **Uni-Variate Bayesian Model**, and another for a **Multi-Variate Bayesian Model**.

```python
def bayesian_model(predictor, response):
    
    X = pd.DataFrame(exploringData[predictor]) #Predictor 
    Y = pd.DataFrame(exploringData[response]) #Response

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, test_size = 200, train_size = 600)  #Split the Dataset into Train and Test
 
    #------------------------------------------------------------------------------------------------
    #BAYESIAN REGRESSION MODEL
    
    model = BayesianRidge()
    model.fit(X_train, Y_train)

    print('Bayesian Intercept \t: b = ', model.intercept_)
    print('Bayesian Coefficients \t: a = ', model.coef_)
   
    regline_x = X_train
    regline_y = model.intercept_ + model.coef_ * X_train #Formula for the Regression line
    
    Y_train_pred = model.predict(X_train) #Predict response variables based on train predictor
    
    print("\nBayesian Train Set:")
    print("Bayesian Explained Variance (R^2):", r2_score(Y_train, Y_train_pred))
    mse = mean_squared_error(Y_train, Y_train_pred)
    print("Mean Squared Error (MSE) \t:", mse)
    print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse)) #Explained Variance (R^2)

    Y_test_pred = model.predict(X_test)

    print("\nBayesian Test Set:")
    print("Bayesian Explained Variance (R^2):", r2_score(Y_test, Y_test_pred)) # Explained Variance (R^2)

    mse = mean_squared_error(Y_test, Y_test_pred)
    print("Mean Squared Error (MSE) \t:", mse)
    print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
    
    # Plot the Predictions vs the True values
    f, axes = plt.subplots(1, 2, figsize=(24, 12))
    axes[0].scatter(X_train, Y_train, color = "blue")
    axes[0].plot(regline_x, regline_y, 'r-', linewidth = 1)
    axes[0].set_xlabel("True values (Train)")
    axes[0].set_ylabel("Predicted values (Train)")
    axes[1].scatter(X_test, Y_test , color = "green")
    axes[1].plot(regline_x, regline_y, 'r-', linewidth = 1)
    axes[1].set_xlabel("True values (Test)")
    axes[1].set_ylabel("Predicted values(Test)")
    plt.show()
```

```python
def multi_variate_bayesian(predictor1, predictor2, response):
    
    X = pd.DataFrame(exploringData[[predictor1, predictor2]]) #Predictor 
    Y = pd.DataFrame(exploringData[response]) #Response

    # Split the Dataset into Train and Test
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, test_size=200, train_size=600)
    
    #------------------------------------------------------------------------------------------------
    #BAYESIAN MODEL
    
    print("---------------------- BAYESIAN MODEL ----------------------")
    model = BayesianRidge()
    model.fit(X_train, Y_train)

    print('Bayesian Intercept \t: b = ', model.intercept_)
    print('Bayesian Coefficients \t: a = ', model.coef_)
    
    Y_train_pred = model.predict(X_train) #Predict
    
    print("\nBayesian Train Set:")
    print("Bayesian Explained Variance (R^2):", r2_score(Y_train, Y_train_pred))
    mse = mean_squared_error(Y_train, Y_train_pred)
    print("Mean Squared Error (MSE) \t:", mse)
    print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse)) #Explained Variance (R^2)

    Y_test_pred = model.predict(X_test)

    print("\nBayesian Test Set:")
    print("Bayesian Explained Variance (R^2):", r2_score(Y_test, Y_test_pred)) # Explained Variance (R^2)

    mse = mean_squared_error(Y_test, Y_test_pred)
    print("Mean Squared Error (MSE) \t:", mse)
    print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
    
    # Plot the Predictions vs the True values
    f, axes = plt.subplots(1, 2, figsize=(24, 12))
    axes[0].scatter(Y_train, Y_train_pred, color = "blue")
    axes[0].plot(Y_train, Y_train, 'r-', linewidth = 1)
    axes[0].set_xlabel("True values of the Response Variable (Train)")
    axes[0].set_ylabel("Predicted values of the Response Variable (Train)")
    axes[1].scatter(Y_test, Y_test_pred, color = "green")
    axes[1].plot(Y_test, Y_test, 'r-', linewidth = 1)
    axes[1].set_xlabel("True values of the Response Variable (Test)")
    axes[1].set_ylabel("Predicted values of the Response Variable (Test)")
    plt.show()
    
    print("Prediction of", response)
    print("Let's predict the value of", response, "for a few specific movies:\nSolo: A Star Wars Story, Kingsman: The Golden Circle and Spirited Away\nThis will be done using the Regression Model derived above.")

    # Extract movies for Prediction
    response_pred = exploringData[exploringData["title"].isin(["Solo: A Star Wars Story", "Kingsman: The Golden Circle", "Spirited Away"])]
    predictors = [predictor1, predictor2]
    # Extract Predictors for Prediction
    X_pred = pd.DataFrame(response_pred[predictors])

    # Predict Response corresponding to Predictors
    y_pred = model.predict(X_pred)

    # Summarize the Actuals, Predictions and Errors
    y_pred = pd.DataFrame(y_pred, columns = ["Predicted"], index = response_pred.index)
    moviedata_acc = pd.concat([response_pred[["title", response]], y_pred], axis = 1)

    y_errs = 100 * abs(moviedata_acc[response] - moviedata_acc["Predicted"]) / moviedata_acc[response]
    y_errs = pd.DataFrame(y_errs, columns = ["Error"], index = response_pred.index)
    moviedata_acc = pd.concat([moviedata_acc, y_errs], axis = 1)
    
    display(moviedata_acc)

    MSE_train = mean_squared_error(Y_train, Y_train_pred)
    StdE_pred = np.sqrt(len(Y_train) * MSE_train/(len(Y_train) - 2))

    print("Mean Squared Error (MSE) \t:", MSE_train.round(2))
    print("Pred Standard Error (SE) \t:", StdE_pred.round(2))
    print("\n")
    
    return model
```
